–í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ

–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ
–í—ã–≤–µ–¥–∏—Ç–µ –¥–≤–∞ —á–∏—Å–ª–∞ ‚Äî —Ä–µ–∑—É–ª—å—Ç–∞—Ç –≤—Ä–∞—â–µ–Ω–∏—è. –û—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–∞—è –ø–æ–≥—Ä–µ—à–Ω–æ—Å—Ç—å –≤–∞—à–µ–≥–æ –æ—Ç–≤–µ—Ç–∞ –Ω–µ –¥–æ–ª–∂–Ω–∞ –ø—Ä–µ–≤—ã—à–∞—Ç—å 10‚Äâ-‚Äâ1.

–ü—Ä–∏–º–µ—Ä—ã\
–≤—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ\
90\
1 1\
–≤—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ\
-1.00000000 1.00000000\
–≤—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ\
180\
1 1390\
–≤—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ\
-1.00000000 -1390.00000000



```python
from math import pi,sin,cos

def rotate_point(x,y,angle):
    fi=angle*pi/180
    nx = round(x*cos(fi)-y*sin(fi), 8)
    ny = round(x*sin(fi)+y*cos(fi), 8)
    return nx,ny


angle=float(input())
x,y=map(float,(input().split()))

print(rotate_point(x,y,angle))
```

–ß–∏—Å–ª–æ –≥—Ä—É—Å—Ç–Ω–æ–µ (upset), –µ—Å–ª–∏ –æ–Ω–æ –Ω–µ –¥–µ–ª–∏—Ç—Å—è –Ω–∏ –Ω–∞ –æ–¥–Ω—É —Å–≤–æ—é —Ü–∏—Ñ—Ä—É.
–ì–µ–Ω–µ—Ä–∞–ª –°—É—Ä–µ–Ω –ø—Ä–æ—Å–∏—Ç –≤–∞—Å –æ–ø—Ä–µ–¥–µ–ª–∏—Ç—å —Ç–∏–ø –∑–∞–¥–∞–Ω–Ω–æ–≥–æ —á–∏—Å–ª–∞.
\
–í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ\

–≤—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ\

```python
def is_happy(num):
    if num<=0:
        return "–ß—Ç–æ-—Ç–æ –Ω–µ —Ç–∞–∫ —Å –≤—Ö–æ–¥–Ω—ã–º–∏ –¥–∞–Ω–Ω—ã–º–∏"
    n = num
    count = 0
    while n > 0:
        a = n%10
        if num%a == 0:
            count+=1
        n = int(n/10)
    if count == 0:
        return "upset"
    if count == len(str(num)):
        return "happier"
    return "happy"


print(is_happy(int(input())))
```

```python
def suitable_teacher(teachers, lecs_in_marker):
    for i in range(len(teachers)):
        if teachers[i] <= lecs_in_marker and teachers[i] > 0:
            return i
    return False

def correct_teacher_order(teachers):
    count_markers = 1
    lecs_in_marker = 0
    res = []
    for i in range(len(teachers)):
        suitable_teacher_ind = suitable_teacher(teachers, lecs_in_marker)
        if not suitable_teacher_ind:
            suitable_teacher_ind = teachers.index(max(teachers))
#         print(suitable_teacher_ind + 1, teachers[suitable_teacher_ind])
        if(teachers[suitable_teacher_ind] <= lecs_in_marker):
            count_markers+=1
            lecs_in_marker = 0
#             print()
        res.append(suitable_teacher_ind + 1)
        teachers[suitable_teacher_ind] = -1 #–ü–æ–º–µ—á–∞–µ–º, —á—Ç–æ —ç—Ç–æ—Ç —Ç–∏—Ç—á–µ—Ä —É–∂–µ –∑–∞—é–∑–∞–Ω
        lecs_in_marker+=1
    res.append(count_markers)

    return res
error = False
n = int(input("–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –Ω–∞–±–æ—Ä–æ–≤ "))
out = ""
for i in range(n):
    count = int(input("–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ª–µ–∫—Ç–æ—Ä–æ–≤ " + str(i+1) + "-–π –Ω–∞–±–æ—Ä "))
    teachers = list(map(int,(input("–ó–Ω–∞—á–µ–Ω–∏—è –ø—Ä–∏–µ–º–ª–µ–º–æ—Å—Ç–∏ " + str(count) + " –ª–µ–∫—Ç–æ—Ä–æ–≤ (—á–µ—Ä–µ–∑ –ø—Ä–æ–±–µ–ª)").split())))
    if len(teachers) != count:
        print("–û—à–∏–±–∫–∞ –ø—Ä–∏ –≤–≤–æ–¥–µ")
        error = True
        break
#     out += " ".join(map(str, correct_teacher_order(teachers))) + "\n"
    res = correct_teacher_order(teachers)
    out += ' '.join(str(res[v]) for v in range(len(res)-1) )
    out += " - –ö–æ–ª-–≤–æ –º–∞—Ä–∫–µ—Ä–æ–≤: " + str(res[len(res)-1]) +"\n"
print()
if not error:
    print(out)
```

?111?1??1\
1.0\
–≤—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ\
1.00000

```python
import random

def task8(memories, chance):
    if(chance > 1):
        print("–û—à–∏–±–∫–∞ –ø—Ä–∏ –≤–≤–æ–¥–µ")
        return
    summ = 0
    for i in range(len(memories)):
        if(memories[i] == '?'):
            summ += int(random.choices([1, 0], weights=[chance, 1 - chance])[0])
        else:
            summ += int(memories[i])
    return summ / len(memories)
m = input()
p = float(input())


print(task8(m,p))
```

## –ó–∞–¥–∞—á–∞ 10
–í–∞–º –¥–∞–Ω–æ –∫–≤–∞–¥—Ä–∞—Ç–Ω–æ–µ —É—Ä–∞–≤–Ω–µ–Ω–∏–µ —Å —Ü–µ–ª—ã–º–∏ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–∞–º–∏ A‚Äâ*‚ÄâX2‚Äâ+‚ÄâB‚Äâ*‚ÄâX‚Äâ+‚ÄâC‚Äâ=‚Äâ0. –ì–∞—Ä–∞–Ω—Ç–∏—Ä—É–µ—Ç—Å—è, —á—Ç–æ A‚Äâ‚â†‚Äâ0 –∏ —É—Ä–∞–≤–Ω–µ–Ω–∏–µ –∏–º–µ–µ—Ç —Ö–æ—Ç—è –±—ã –æ–¥–∏–Ω –¥–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω—ã–π –∫–æ—Ä–µ–Ω—å. –ù–∞–π–¥–∏—Ç–µ –∫–æ—Ä–Ω–∏ —É—Ä–∞–≤–Ω–µ–Ω–∏—è.\

–í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ\
–ï–¥–∏–Ω—Å—Ç–≤–µ–Ω–Ω–∞—è —Å—Ç—Ä–æ–∫–∞ –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö —Å–æ–¥–µ—Ä–∂–∏—Ç —Ü–µ–ª—ã–µ —á–∏—Å–ª–∞ A,‚ÄâB –∏ C (‚Äâ-‚Äâ1000‚Äâ‚â§‚ÄâA,‚ÄâB,‚ÄâC‚Äâ‚â§‚Äâ1000,‚ÄâA‚Äâ‚â†‚Äâ0).\

–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ\
–í—ã–≤–µ–¥–∏—Ç–µ –∫–æ—Ä–Ω–∏ —É—Ä–∞–≤–Ω–µ–Ω–∏—è –≤ –ø–æ—Ä—è–¥–∫–µ –∏—Ö –≤–æ–∑—Ä–∞—Å—Ç–∞–Ω–∏—è. –ï—Å–ª–∏ —É—Ä–∞–≤–Ω–µ–Ω–∏–µ –∏–º–µ–µ—Ç –æ–¥–∏–Ω –∫–æ—Ä–µ–Ω—å –∫—Ä–∞—Ç–Ω–æ—Å—Ç–∏ 2, –≤—ã–≤–µ–¥–∏—Ç–µ –µ–≥–æ —Ç–æ–ª—å–∫–æ –æ–¥–∏–Ω —Ä–∞–∑. –ö–æ—Ä–µ–Ω—å —Å—á–∏—Ç–∞–µ—Ç—Å—è –ø—Ä–∞–≤–∏–ª—å–Ω—ã–º, –µ—Å–ª–∏ –µ–≥–æ –∞–±—Å–æ–ª—é—Ç–Ω–∞—è –∏–ª–∏ –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–∞—è –ø–æ–≥—Ä–µ—à–Ω–æ—Å—Ç—å –Ω–µ –ø—Ä–µ–≤–æ—Å—Ö–æ–¥–∏—Ç 10‚Äâ-‚Äâ4.\
\
–ü—Ä–∏–º–µ—Ä—ã\
–≤—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ\
1 -2 1\
–≤—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ\
1\
–≤—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ\
1 0 -1\
–≤—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ\
-1 1\

```python
import math
def quadraticsolver(a,b,c) :
    D = b ** 2 - 4 * a * c

    if D == 0:
        x1 = (-b - math.sqrt(D)) / (2*a)
        return x1, x1
    elif D > 0:
        x1 = (-b - math.sqrt(D)) / (2*a)
        x2 = (-b + math.sqrt(D)) / (2*a)

        if x1 > x2:
            return x2, x1
        else:
            return x1, x2

a,b,c = map(float,(input().split()))
x1, x2 = quadraticsolver(a,b,c)
if x1 == x2:
    print(x1)
else:
    print(x1, x2)
```
